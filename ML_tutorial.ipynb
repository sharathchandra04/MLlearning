```python
a = 5

import numpy as np
import matplotlib.pyplot as plt

# Generate sample data for features and target
np.random.seed(0)
X = 2 * np.random.rand(100, 2)  # Two features
y = 3 + 4 * X[:, 0] + 2 * X[:, 1] + np.random.randn(100) * 2  # Linear relation with some noise

# Define the range of coefficients (theta0 and theta1)
theta0_vals = np.linspace(-10, 10, 100)
theta1_vals = np.linspace(-10, 10, 100)

# Create a mesh grid for the coefficients
theta0, theta1 = np.meshgrid(theta0_vals, theta1_vals)

# Calculate RSS for each combination of coefficients
rss = np.zeros_like(theta0)
for i in range(len(theta0_vals)):
    for j in range(len(theta1_vals)):
        y_pred = theta0[i, j] + theta1[i, j] * X[:, 0]
        rss[i, j] = np.sum((y_pred - y) ** 2)

# Plot the contour
plt.contour(theta0, theta1, rss, levels=20, cmap='viridis')
plt.xlabel('Theta0 (Intercept)')
plt.ylabel('Theta1 (Coefficient for Feature 1)')
plt.title('RSS Contours for Linear Regression Coefficients')
plt.colorbar(label='RSS')
plt.show()
```
